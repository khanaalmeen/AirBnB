## Univariant Analysis

**Room Types Column**

```python
df.room_type.value_counts()
```

```python
plt.figure(figsize=(9,6))
sns.countplot(data=df, x = 'room_type')
plt.show()
```

```python
plt.figure(figsize=(9,6))
df.room_type.value_counts().plot(kind='pie', autopct='%1.2f%%')
plt.show()
```
